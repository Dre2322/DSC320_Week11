```python
# 1. Using the Binomial Distribution
# a

from scipy.stats import binom

def ensemble_accuracy(N, p):
  """Calculates the accuracy of an ensemble model

  Args:
    N: number of models
    p: accuracy of each model

  Returns:
    The accuracy of the ensemble model
  """

  return 1 - binom.cdf(N // 2, N, p)

# Calculate the accuracy of the ensemble model
accuracy = ensemble_accuracy(15, 0.63)

# Print the accuracy
print(f"Accuracy of the ensemble: {accuracy *100} %")
```

```python
# 2. Working with Random Data
# a

import random

# Generate a list of 10 random numbers between 1 and 5
random_numbers = [random.randint(1,5) for i in range(10)]

# Create a dictionary to store the pmf
pmf = {}
for number in random_numbers:
    if number not in pmf:
        pmf[number] = 0
    pmf[number] += 1

# Calculate the probability of each number
total_numbers = len(random_numbers)  # Total number of generated numbers
for number in pmf:
    pmf[number] /= total_numbers

# Print the PMF
print(pmf)
```

```python
# c

# find the mean 1000 times and create a list of the means
means = []
for i in range(1000):
  random_numbers = [random.randint(1,5) for i in range(50)]
  pmf = {}
  for number in random_numbers:
    if number not in pmf:
        pmf[number] = 0
    pmf[number] += 1
  total_numbers = len(random_numbers)
  for number in pmf:
      pmf[number] /= total_numbers
  mean = sum(number * pmf[number] for number in pmf) / sum(pmf.values())
  means.append(mean)

# print the list of means
print(means)
```

```python
# d

import matplotlib.pylab as plt

# create a histogram of the means
plt.hist(means, bins=10)
plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.title('Histogram of means')
plt.show()
```

```python
# e

# find mean of list of means
mean_of_means = sum(means) / len(means)

# find standard deviation of list of means
std_dev = (sum((x - mean_of_means) ** 2 for x in means) / len(means)) ** 0.5

# print the mean and standard deviation
print("Mean:", mean_of_means)
print("Standard Deviation:", std_dev)
```
